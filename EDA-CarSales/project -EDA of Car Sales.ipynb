### 3.3 Preprocessing

- Dealing with missing,Zeros,Null and duplicates <br/>
    - Dropping zero value of __price.__
    - Replacing missing values of __engV__ with mean.
    - Dropping engV which is more than or equal to 9 as it is incorrect value.     
    - Replacing null values of column __drive__ with mode.
    - Droping price with zeros as is only 2.5 of the total data
    - dropping duplicates - 113 columns  

_Dropping_ _113_ _Duplicate_ _Rows._

```python
cs_data = cs_data.drop_duplicates(keep = 'first')  # dropping duplicates - 113 rows
```

```python
cs_data.describe()
```

_Check number_ of cars with __engV__ more than __8.__ _Based on the volume we can decide whether to delete or replace with mean value._

```python
# Check the number of cars with engv more than 8. 
cs_data.loc[cs_data["engV"] > 8].count()
```

_As the count is 62 and it is less than 1% of total volume of the dataset, we will drop those columns._

```python
# drop engV which is more than 8 (62 entries)
cs_drop_engV = cs_data[ cs_data['engV'] > 8 ].index
cs_data.drop(cs_drop_engV, inplace = True)
```

_Replacing Null values of column __drive__ with mode._

```python
#Replacing null values of column __drive__ with mode.
cs_data.drive = cs_data.drive.fillna(cs_data["drive"].mode()[0])
```

_Replacing missing values of __engV__ with mean._

```python
mean_engV = cs_data.engV.mean()                            # replacing missing values of engV with mean
cs_data.engV.fillna(mean_engV,inplace = True)
```

_Replacing zero values of __mileage__ with mean._

```python
cs_data['mileage']=cs_data['mileage'].replace(0,cs_data['mileage'].mean())   # replace zeros in milage with mean
```

_Droping __price__ with zeros as is only 2.5 of the total data._

```python
cs_drop_price = cs_data[cs_data['price'] == 0].index     # droping price with zeros as is only 2.5 of the total data
cs_data.drop(cs_drop_price,inplace = True)
```

_Cleaned dateset file created as a __csv__ file with name __cs-cleaned-data.__

```python
cs_data.to_csv('cs-cleaned-data',index = False)   # cleaned dataset file created as csv file.
```
